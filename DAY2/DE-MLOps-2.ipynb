```python
zscore
=======
 |-> standard score

     X - mean of the dataset
Z=    --------------------------
         standard deviation

Zscore
```

```python
d = np.array([50,60,70,80,90])
mean = np.mean(d)
std_dev = np.std(d)
zscore = (d - mean) / std_dev
zscore
```

```python
d[np.abs(zscore) >2]
```

```python
plt.boxplot(d)
```

```python
d = np.array([50,60,70,80,90,100,110,150,1200,160])
mean = np.mean(d)
std_dev = np.std(d)
zscore = (d - mean) / std_dev
print(zscore)
d[np.abs(zscore) > 2]
```
